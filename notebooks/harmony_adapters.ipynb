```python
from pathlib import Path

from harmonykit.adapter import BaseAdapter

# Point the adapter at the dataset root (contains the "harmonies" folder)
adapter = BaseAdapter(
    dataset_name="Mozart Piano Sonatas",
    root_dir=Path("../data/mozart_piano_sonatas"),
)

# Load and display every HarmonyEvent for the K545-1 piece
events = list(adapter.events("K545-1"))
for event in events:
    print(event)
```
